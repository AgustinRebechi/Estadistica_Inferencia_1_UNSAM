 Paracada ùëõ entre 1 y 3000,generar observaciones $x_1 ,..., x_n$ de $X_1,. . .,X_n$ v.a iid con distribuci√≥n exponencial con par√°metro ùúÜ = 1/2, con ùúÜ a elecci√≥n. Luego, obtener ¬Ø ùë•ùëõ como la media de estas observaciones, es decir: $$\widehat{x}_n = \frac{1}{n} \sum_{i=1}^{n} X_i  $$
 Graficar $n\ vs \ \widehat{x}_n $ e interpretar

## Desarrollo

Importo la librer√≠a Scipy e inicializo la funci√≥n exponencial con $\lambda = 1/2‚ÄØ$

```python
from scipy.stats import expon

exponencial = expon(scale=2)
```

Defino una funci√≥n que tome como argumento *n* y que devuelva la v.a con esas en observaciones:

```python
def media_n(n: int, distribucion) -> float:
    # Genero en valores aleatorios de la distribuci√≥n
    valores = distribucion.rvs(size=n)

    # Calculo la media muestral
    media = sum(valores) / n

    return media
```

Genero las 3000 pruebas y las voy guardando en una lista para luego graficar

```python
medias_muestrales = [media_n(n, exponencial) for n in range(1,30000)]
```

Ahora finalmente genero el gr√°fico

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.plot(range(1,30000), medias_muestrales, label="Media Muestral")
plt.axhline(y=2,color="r",linestyle="--",label="Valor Esperado (2)")
plt.xlabel("n")
plt.ylabel("Media Muestral")
plt.title("n vs Media Muestral")
plt.legend()
plt.grid(True)
plt.show()
```
